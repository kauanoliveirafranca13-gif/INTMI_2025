```python
soma_idades = 0
quantidade_idades = 0

print("Digite as idades (digite um número negativo para finalizar):")

while True:
    try:
        idade = int(input("Idade: "))
        if idade < 0:
            break

        soma_idades += idade
        quantidade_idades += 1
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if quantidade_idades > 0:
    media_idades = soma_idades / quantidade_idades
    print(f"A média das idades digitadas é: {media_idades:.2f}")
else:
    print("Nenhuma idade válida foi digitada para calcular a média.")
```

```python
positivos = 0
negativos = 0
zeros = 0

print("Digite 10 números:")
for i in range(1, 11):
    try:
        numero = float(input(f"Número {i}: "))
        if numero > 0:
            positivos += 1
        elif numero < 0:
            negativos += 1
        else:
            zeros += 1
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")

print(f"Resultados da análise:\nPositivos: {positivos}\nNegativos: {negativos}\nZeros: {zeros}")
```

```python
senha_correta = "secreta123"
max_tentativas = 3

print(f"Sistema de Login (você tem {max_tentativas} tentativas):")

for tentativa in range(1, max_tentativas + 1):
    senha_digitada = input(f"Tentativa {tentativa}/{max_tentativas} - Digite a senha: ")

    if senha_digitada == senha_correta:
        print("Bem-vindo! Login realizado com sucesso.")
        break # Sai do laço se a senha estiver correta
    else:
        print("Senha incorreta. Tente novamente.")

# Se o laço terminar e a senha não foi correta (ou seja, o break não foi acionado)
if senha_digitada != senha_correta:
    print("Número máximo de tentativas excedido. Acesso bloqueado.")
```

```python
soma_positivos = 0

print("Digite 10 números:")
for i in range(1, 11):
    try:
        numero = float(input(f"Número {i}: "))
        if numero < 0:
            print(f"Número {numero} é negativo e será ignorado.")
            continue  # Ignora números negativos
        soma_positivos += numero
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")

print(f"A soma dos números positivos digitados é: {soma_positivos}")
```

```python
soma_acumulada = 0
numero_atual = 1

print("Iniciando acumulador (parará quando a soma ultrapassar 200):")

while True:
    soma_acumulada += numero_atual
    print(f"Adicionando {numero_atual}, Soma atual: {soma_acumulada}")

    if soma_acumulada > 200:
        print("Soma ultrapassou 200. Encerrando.")
        break

    numero_atual += 1

print(f"Acumulador finalizado. Soma total: {soma_acumulada}")
```

```python
contador_palavras = 0

print("Digite palavras (digite 'fim' para encerrar):")

while True:
    palavra = input("Palavra: ")

    if palavra.lower() == "fim":
        print("Palavra 'fim' detectada. Encerrando contagem.")
        break

    contador_palavras += 1
    print(f"'{palavra}' registrada. Total de palavras: {contador_palavras}")

print(f"\nTotal de palavras válidas digitadas: {contador_palavras}")
```
